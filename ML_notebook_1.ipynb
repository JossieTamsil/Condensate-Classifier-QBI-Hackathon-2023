```
from imblearn.combine import SMOTEENN
smote_enn = SMOTEENN(random_state=0)
X_resampled, y_resampled = smote_enn.fit_resample(X_train, y_train)
```

### normalize expression level

```
normalizer = preprocessing.Normalizer().fit(X_resampled)
```

### Logistic Regression

```
from sklearn.linear_model import LogisticRegression

lf = LogisticRegression(random_state=0, multi_class='multinomial').fit(normalizer.transform(X_resampled), y_resampled)
```

```
print(classification_report(y_resampled, lf.predict(normalizer.transform(X_resampled))))
```

```
from sklearn.metrics import classification_report
print(classification_report(y_test, lf.predict(normalizer.transform(X_test))))
```

### SVC

```
from sklearn.svm import SVC

svc = SVC().fit(normalizer.transform(X_resampled), y_resampled)
```

```
print(classification_report(y_test, svc.predict(normalizer.transform(X_test))))
```

### Random Forest

```
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier().fit(normalizer.transform(X_resampled), y_resampled)

print(classification_report(y_test, rf.predict(normalizer.transform(X_test))))
```
